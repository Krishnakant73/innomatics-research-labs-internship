<u>Q6)Write a function</u>

```python
def is_leap(year):
    leap = False
    
    if year % 400 == 0:
        leap = True
    elif year % 100 == 0:
        leap = False
    elif year % 4 == 0:
        leap = True
    return leap

year = int(input())
print(is_leap(year))
```

<u>Q1)Running Sum of 1d Array </u>

```python
class Solution:
    def runningsum(self, nums):
        result = []
        running_sum = 0
        
        for i in nums:
            running_sum += i       
            result.append(running_sum)  
        return result

```

```python
nums = [1,2,3,4]
print(Solution().runningsum(nums))
```

<u>Q2)Shuffle the Array</u>

```python
class Solution:
    def shuffle(self, nums, n):
        result = []
        for i in range(n):
            result.append(nums[i])     
            result.append(nums[i+n])   
        return result
```

```python
nums = [2,5,1,3,4,7]
n = 3
print(Solution() .shuffle(nums, n))
```

<u>Q3)Kids With the Greatest Number of Candies</u>

```python
class Solution:
    def kidswithcandies(self, candies, extracandies):
        max_candies = max(candies)
        result = []
        for c in candies:
            result.append(c + extracandies >= max_candies)
        return result
```

```python
candies = [4,2,1,1,2]
extracandies = 1
print(Solution().kidswithcandies(candies, extracandies))
```
